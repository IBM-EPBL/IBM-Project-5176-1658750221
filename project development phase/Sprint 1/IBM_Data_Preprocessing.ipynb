### Visualizing And Preprocessing Dataset

```python
dataset_path = '/content/drive/MyDrive/IBM/Dataset/TRAIN_SET'
class_names = ['APPLES', 'BANANA', 'ORANGE', 'PINEAPPLE','WATERMELON']
```

```python
f, axs = plt.subplots(1, 5, figsize=(15,15))
for i in range(5):
  fruit_path = os.path.join(dataset_path, class_names[i], '*')
  fruit_path = glob.glob(fruit_path)
  image = io.imread(fruit_path[10])

  axs[i].imshow(image)
```

```python
# apply glob module to retrieve files/pathnames  

fruit_path = os.path.join(dataset_path, class_names[0], '*')
fruit_path = glob.glob(fruit_path)
```

```python
# accessing an image file from the dataset classes
image = io.imread(fruit_path[10])  

# plotting the original image
i, (im1) = plt.subplots(1)
i.set_figwidth(15)
im1.imshow(image)
```

```python
# plotting the original image and the RGB channels  

i, (im1, im2, im3, im4) = plt.subplots(1, 4, sharey=True)
i.set_figwidth(20) 

im1.imshow(image)  #Original image
im2.imshow(image[:, : , 0]) #Red
im3.imshow(image[:, : , 1]) #Green
im4.imshow(image[:, : , 2]) #Blue
i.suptitle('Original & RGB image channels')
```

```python
import skimage.io
import skimage.color

gray_image = skimage.color.rgb2gray(image)
plt.imshow(gray_image, cmap = 'gray')
```

```python
norm_image = (gray_image - np.min(gray_image)) / (np.max(gray_image) - np.min(gray_image))
plt.imshow(norm_image)
```
